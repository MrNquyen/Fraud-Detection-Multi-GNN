```python
import pandas as pd
import numpy as np
```

```python
path = "F:\\UNIVERSITY\\UNIVERSITY_DOCUMENTS\\BIG_DATA\\Doan\\baseline\\multi_gnn\\Multi-GNN\\data\\Small_HI\\formatted_transactions.csv"
df = pd.read_csv(path)
```

```python
df.columns
```

```python
df[df["Timestamp"] > 10]
```

```python
list_daily_df = []
num_record_selected = 1000
# Select 10 days
for day in range(10):
    l = day * 24 * 3600
    r = (day + 1) * 24 * 3600
    daily_df = df.loc[
        (df["Timestamp"] >= l) & (df["Timestamp"] < r)
    ].iloc[:num_record_selected]
    list_daily_df.append(daily_df)

sample_df = pd.concat(list_daily_df)
```

```python
save_path = "data/sample/formatted_transactions.csv"
sample_df.to_csv(save_path)
```
